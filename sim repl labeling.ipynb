```python
# Read the emails file
import pandas as pd

dataset = pd.read_excel("output_tfidf_clustering_with_custom_keywords.xlsx")
```

```python
pd.set_option('display.max_columns', None)
dataset.describe()
```

```python
# (dataset['replace my old sim'] == 1) |
sim_repl_dataset = dataset[
                           (dataset['replace sim'] == 1) | 
                           (dataset['my existing sim'] == 1) |
                           (dataset['existing sim'] == 1) | 
                           (dataset['my old sim'] == 1) |
                           (dataset['old sim'] == 1)|
                           (dataset['replacement of sim'] == 1) | 
                           (dataset['replace my old sim card'] == 1) |
                           (dataset['replace old sim'] == 1) |
                           (dataset['replacement sim'] == 1) |
                           (dataset['replacement sim card'] == 1) |
                           (dataset['sim replacement'] == 1)
                          ]
```
